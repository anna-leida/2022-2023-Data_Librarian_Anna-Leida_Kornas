```python
import urllib.request
import pandas as pd
```

```python
#Checkout-Daten über die API beziehen und als CSV-Datei speichern
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2018&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
#In einen DataFrame laden und analysieren
checkouts_df = pd.read_csv(checkout_file)
checkouts_df
```

```python
#Tabelle nur mit den Filmen in movie speichern
movie = checkouts_df[checkouts_df["materialtype"] == "MOVIE"]
movie
```

```python
#Wie viele Titel sind in der Tabelle eintahlten?
movie["title"].count()
```

```python
#Welche Genres haben die Filme? 
genres = movie["subjects"].sum()
genres
```

```python
#Wie oft ist das Genre "Doucumentary" vorhanden?
genres.count("Documentary")
```

```python
#Wie oft ist das Genre "Drama" vorhanden?
genres.count("Drama")
```

```python
#geordnete Auflistung der Genres
genres_movie = movie["subjects"]
genres_movie.unique()
```

```python
checkouts = movie["checkouts"]
checkouts
```

```python
#Übersicht der Ausleihen
checkouts.hist(grid=False, color="#206020", legend=True)
```
